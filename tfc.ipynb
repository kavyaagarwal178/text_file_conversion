```python
# text to file conversion
f=open('th.txt','w')
f.write('a'*200*1024*256)
f=open('fh.txt','w')
f.write('a'*400*1024*1256)
f=open('sh.txt','w')
f.write('a'*600*1024*256)
f=open('eh.txt','w')
f.write('a'*800*1024*256)
f=open('thh.txt','w')
f.write('a'*1000*1024*256)

```

```python
import time
timeTaken=[]


begin1=time.time()
fo=open('th.txt','r')
for x in fo.read():
  y=x.upper()
fo1=open('th1.txt','w')
fo1.write(y)
end1=time.time()
timeTaken.append(end1-begin1)

begin1=time.time()
fo=open('fh.txt','r')
for x in fo.read():
  y=x.upper()
fo1=open('fh1.txt','w')
fo1.write(y)
end1=time.time()
timeTaken.append(end1-begin1)

begin1=time.time()
fo=open('sh.txt','r')
for x in fo.read():
  y=x.upper()
fo1=open('sh1.txt','w')
fo1.write(y)
end1=time.time()
timeTaken.append(end1-begin1)

begin1=time.time()
fo=open('eh.txt','r')
for x in fo.read():
  y=x.upper()
fo1=open('eh1.txt','w')
fo1.write(y)
end1=time.time()
timeTaken.append(end1-begin1)

begin1=time.time()
fo=open('thh.txt','r')
for x in fo.read():
  y=x.upper()
fo1=open('thh1.txt','w')
fo1.write(y)
end1=time.time()
timeTaken.append(end1-begin1)


```

```python
print(timeTaken)
```

```python
import matplotlib.pyplot as plt
size=[200*1024*256,400*1024*256,600*1024*256,800*1024*256,1000*1024*256]
plt.plot(size,timeTaken)
plt.xlabel(size)
plt.ylabel(timeTaken)
plt.show()
```
